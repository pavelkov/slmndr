```python
import cv2
import numpy as np
```

```python
name = '20220613_123021'
img = cv2.imread(f"/home/hobo/Dropbox/Dev/Slmndr/Data/Images/{name}.jpg")

b = img[:,:,0] # cv2.equalizeHist(img[:,:,0])
g = img[:,:,1] # cv2.equalizeHist(img[:,:,1])
r = img[:,:,2] # cv2.equalizeHist(img[:,:,2])

Y = (r >= 180) & (g >= 120) & (b <= 200)
B = (r <= 100) & (g <= 100) & (b <= 100)

cv2.imwrite(f"/home/hobo/Dropbox/Dev/Slmndr/Data/Temp/{name}.png", np.dstack((255 * Y, 255 * B, np.zeros_like(Y))))
```
